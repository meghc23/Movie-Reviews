```python
all_words = nltk.FreqDist(movie_reviews.words())
feature_vector = list(all_words)[:4000]
```

```python
# Initialization
feature = {}
# One movie review is chosen
review = movie_reviews.words('neg/cv954_19932.txt')
# True is assigned if word in feature_vector can also be found in review. Otherwise False
for x in range(len(feature_vector)):
 feature[feature_vector[x]] = feature_vector[x] in review
# The words which are assigned True are printed
[x for x in feature_vector if feature[x] == True]
```

```python
document = [(movie_reviews.words(file_id),category) for file_id in movie_reviews.fileids() for category in movie_reviews.categories(file_id)]
document
```

```python
# we define a function that finds the features
def find_feature(word_list):
# Initialization
  feature = {}
# For loop to find the feature. True is assigned if word in feature_vector can also be found in review. Otherwise False
  for x in feature_vector:
    feature[x] = x in word_list
  return feature
```

```python
feature_sets = [(find_feature(word_list),category) for (word_list,category) in document]
```

```python
import pandas as pd
df = pd.DataFrame(document , columns= ['reviews' , 'tag'])
```

```python
df.head()
```

```python
df['tag'].value_counts()
```

```python
df['reviews'][0]
```

```python
df['reviews'] = df['reviews'].apply(lambda x : '  '.join(x))
```

```python
df.head()
```

```python
df['reviews']=df['reviews'].str.replace('@[a-zA-Z]*','')
```

```python
df['reviews'] = df['reviews'].apply(lambda x : re.sub('[^a-zA-Z#\s]*','',x))
```

```python
stop_words = set(stopwords.words('english'))
df['reviews'] = df['reviews'].apply(lambda x : ' '.join([w for w in x.split() if not w in stop_words]))
```

```python
df.head()
```

```python
tweet_len = df['reviews'].str.len()
plt.hist(tweet_len,bins=20,label='train')
plt.show()
```

```python
X = df['reviews']
y = df['tag']
```

```python
dict1 = {'neg':0 , 'pos':1}
y = y.map(dict1)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.25 , stratify = y)
```

```python
from sklearn.feature_extraction.text import CountVectorizer , TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer(max_df = 0.9 , min_df = 2 , stop_words = 'english' , max_features = 1000)
X_train1 = tfidf_vectorizer.fit_transform(X_train)
```

```python
X_test1 = tfidf_vectorizer.transform(X_test)
```

```python
X_train1.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train1, y_train)
pred = lr.predict(X_test1)
```

```python
from sklearn import metrics
metrics.confusion_matrix(pred , y_test)
```

```python
print(metrics.classification_report(pred , y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
log=RandomForestClassifier(random_state=5)
log.fit(X_train1,y_train)
pred2=log.predict(X_test1)
print(metrics.classification_report(y_test,pred2))
print(metrics.confusion_matrix(y_test,pred2))
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train1.toarray(),y_train)
pred7=svc.predict(X_test1.toarray())
print(metrics.classification_report(y_test,pred7))
print(metrics.confusion_matrix(y_test,pred7))
```
